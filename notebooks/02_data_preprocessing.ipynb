```python
# notebooks/02_data_preprocessing.ipynb

import json
import os
from datasets import load_dataset
```

```python
print("--- 2단계: 데이터 준비 시작 ---")

# --- 1. "벌크 데이터" (찐친 말투) 다운로드 ---
# (Krew/KoCulture-Dialogues 데이터셋을 로드합니다)
print("Krew/KoCulture-Dialogues 데이터셋 로드를 시작합니다...")

dataset_dict = load_dataset("huggingface-KREW/KoCulture-Dialogues", revision="refs/convert/parquet")
    
dataset = dataset_dict['train']
print("데이터셋 로드 성공!")
    
```

```python
if 'dataset' in locals():
    output_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'ko_culture.jsonl'))
    
    print(f"데이터를 변환하여 '{output_path}'에 저장합니다...")
    
    count = 0
    # 'dataset' (train 스플릿)에서 데이터를 하나씩 처리
    with open(output_path, 'w', encoding='utf-8') as f:
        for row in dataset:
            # 원본 데이터: {'question': '...', 'answer': '...'}
            user_content = row['question']
            assistant_content = row['answer']
            
            # 우리 형식: {"messages": [{"role": "user", ...}, {"role": "assistant", ...}]}
            formatted_data = {
                "messages": [
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": assistant_content}
                ]
            }
            
            # jsonl 파일로 저장 (한 줄에 하나의 json)
            f.write(json.dumps(formatted_data, ensure_ascii=False) + '\n')
            count += 1

    print(f"--- 변환 완료 ---")
    print(f"총 {count}개의 대화가 {output_path}에 저장되었습니다.")
else:
    print("--- 'dataset' 변수가 로드되지 않아 파일 저장을 스킵합니다 ---")
```
